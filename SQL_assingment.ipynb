## Q1 --- List all the directors who directed a 'Comedy' movie in a leap year. (You need to check that the genre is 'Comedy’ and year is a leap year) Your query should return director name, the movie name, and the year.

<h4>To determine whether a year is a leap year, follow these steps:</h4>

<ul>
    <li><b>STEP-1:</b> If the year is evenly divisible by 4, go to step 2. Otherwise, go to step 5.</li>
    <li><b>STEP-2:</b> If the year is evenly divisible by 100, go to step 3. Otherwise, go to step 4.</li>
    <li><b>STEP-3:</b> If the year is evenly divisible by 400, go to step 4. Otherwise, go to step 5.</li>
    <li><b>STEP-4:</b> The year is a leap year (it has 366 days).</li>
    <li><b>STEP-5:</b> The year is not a leap year (it has 365 days).</li>
</ul>

Year 1900 is divisible by 4 and 100 but it is not divisible by 400, so it is not a leap year.

```python
#https://stackoverflow.com/questions/6534788/check-for-leap-year
#Tables used 
#director name     movie - mid - person
#year              movie year
#comdey genre       movie m_genre genre
```

```python
%%time
def grader_1(q1):
    q1_results  = pd.read_sql_query(q1,conn)
    print(q1_results.head(10))
    assert (q1_results.shape == (232,3))

query1 = """ 
select p.name, m.title,m.year 

from movie as m 

join m_director as md on m.mid = md.mid 

join person as p on p.pid = md.pid 

join  m_genre as mg on m.mid = mg.mid

join genre as g on mg.gid = g.gid

where  year in (select year from movie where (year % 400 =0) or (year % 100 != 0) and (year % 4 =0 ))
AND   g.name in (select name from genre where name like '%Comedy%' )

"""
grader_1(query1)
```

```python
#https://stackoverflow.com/questions/59354472/sql-calculating-percentage-by-joining-a-sub-table-to-another
```

```python
%%time
def grader_5b(q5b):
    q5b_results  = pd.read_sql_query(q5b,conn)
    print(q5b_results.head(10))
    assert (q5b_results.shape == (4,3))

query5b = """ with 
females as (
select m.mid  from movie as m 
join m_cast as mc on m.mid = mc.mid
join person as p on mc.pid = p.pid
group by m.mid, p.gender
having m.mid not in (  select m.mid  from movie as m 
                      join m_cast as mc on m.mid = mc.mid
                      join person as p on mc.pid = p.pid
                  group by m.mid, p.gender
                      having p.gender in  ('Male',NULL) )),

movie_years as (select m.year , count(*) as movie_count
                from movie as m 
                group by 1)



select m.year, count(m.year)*100/my.movie_count as percentageMovies, my.movie_count
from movie as  m
join females as f on f.mid = m.mid
join movie_years as my on  m.year = my.year
group by m.year
"""
grader_5b(query5b)
```

```python
#srk movies id
query = """

with Shahurukh_1 as

( select distinct trim(mc1.pid) pid from ( select distinct(TRIM(mc.mid)) SHARUKH_MOVIE_ID, P.PID PID 
                                    FROM M_CAST MC JOIN PERSON P ON TRIM(MC.PID) = P.PID WHERE TRIM(P.NAME) 
                                    LIKE '%Shah Rukh Khan%' )
                                    T1 JOIN M_CAST MC1 ON TRIM(MC1.MID) = T1.SHARUKH_MOVIE_ID WHERE TRIM(MC1.PID) <> T1.PID ),


SHAHRUKH_2_MOVIES AS ( select DISTINCT(TRIM(MC.MID)) SHARUKH2_MOVIE_ID, SN1A.PID PID 
                      from m_cast as mc join Shahurukh_1 SN1A ON TRIM(mc.pid) = SN1A.PID )

select P.NAME Actor_Name 
from person P
where P.PID IN (select TRIM(MC.PID) from M_CAST MC where TRIM(MC.MID) IN 
                              (select SHARUKH2_MOVIE_ID FROM SHAHRUKH_2_MOVIES)) 
                                and P.PID not in  (select PID from Shahurukh_1) 
                                and P.PID  not in  (select pid from person 
                                                    where name like '%Shah Rukh Khan%')

"""

df = pd.read_sql_query(query,conn)
df
```
