2. Write a program to print twin primes less than 1000. If two consecutive odd numbers are
both prime then they are known as twin primes

```python
def consec_prime(num):
  consec_prime=[]

  def prime(num):  
    '''Function for checking if number is prime or not'''
    for i in range(2,num,):
      if num%i==0:
        break
    else:return num

  prime_nums =[2] # Intialising list 2 then iterating for odd numbers
  for i in range(3,num+1,2):
    if prime(i) == None:
      continue
    else:
      prime_nums.append(prime(i))

  for i in range(len(prime_nums)-1):
    if prime_nums[i+1] - prime_nums[i] ==2:
      consec_prime.append((prime_nums[i],prime_nums[i+1]))
  
  return consec_prime
  
```

3 Write a program to find out the prime factors of a number. Example: prime factors of 56 -
2, 2, 2, 7

```python
#https://www.geeksforgeeks.org/print-all-prime-factors-of-a-given-number/

#Interate for 2 then all odd number including itself
def factors(num):
  factors = []
  while num%2==0: # dividing by two
    factors.append(2)
    num = num/2
  
  for i in range(3,int(num)+1): # looping for odd factors till num
    while num%i==0:
      factors.append(i)
      num = num/i

  return factors
```

4 Write a program to implement these formulae of permutations and combinations.
Number of permutations of n objects taken r at a time: p(n, r) = n! / (n-r)!. Number of
combinations of n objects taken r at a time is: c(n, r) = n! / (r!*(n-r)!) = p(n,r) / r!

```python
def fact(num):
  assert num >= 1 and int(num)==num, 'Factorial not defined'
  if num ==1:return 1
  else: return num*fact(num-1)

def p(n,r):
  return fact(n)/fact(n-r)

def c(n,r):
  return p(n,r)/fact(r)
```

6 Write a function cubesum() that accepts an integer and returns the sum of the cubes of
individual digits of that number. Use this function to make functions PrintArmstrong() and
isArmstrong() to print Armstrong numbers and to find whether is an Armstrong number.

```python
def cubesum(num,power):
  cubes = 0
  while num!=0:
    c = num%10
    cubes += c**power
    num = num//10
  
  return cubes

def isArmstong(num):
  return num == cubesum(num,3) # taking 3 for  digit numbers

def PrintArmstrong(a,b):
  # print armstrong number between a and b
  for i in range(a,b):
    if isArmstong(i):
      print('This is an armstrong number {}'.format(i))


```

```python
PrintArmstrong(100,900)
```

```python
#https://www.geeksforgeeks.org/program-to-calculate-product-of-digits-of-a-number/
# Includeds the case when 0 is in between
def ProdDigits(num):
  prod=1
  while num !=0:
    p = num%10
    if p ==0:
      p=1
    prod *=p
    num //=10
  return prod
```

8 If all digits of a number n are multiplied by each other repeating with the product, the one
digit number obtained at last is called the multiplicative digital root of n. The number of
times digits need to be multiplied to reach one digit is called the multiplicative
persistance of n.
Example: 86 -> 48 -> 32 -> 6 (MDR 6, MPersistence 3)
 341 -> 12->2 (MDR 2, MPersistence 2)
Using the function prodDigits() of previous exercise write functions MDR() and
MPersistence() that input a number and return its multiplicative digital root and
multiplicative persistence respectively

```python
def MDR(num):
  while num//10!=0:
    num = ProdDigits(num)
  return num

def MPersistence(num):
  count = 0
  while num//10!=0:
    num = ProdDigits(num)
    count+=1
  return count
```

```python
print(MDR(86))
print(MDR(341))
```

```python
print(MPersistence(86))
print(MPersistence(341))
```

9 Write a function sumPdivisors() that finds the sum of proper divisors of a number. Proper
divisors of a number are those numbers by which the number is divisible, except the
number itself. For example proper divisors of 36 are 1, 2, 3, 4, 6, 9, 18

```python
# 12 will be also divisor of 36
def SumPdivisors(num):
  divisors =[]
  for i in range(1,num):
    if num%i==0:
      divisors.append(i)
  #print(divisors)
  return sum(divisors)
```

```python
SumPdivisors(36)
```
