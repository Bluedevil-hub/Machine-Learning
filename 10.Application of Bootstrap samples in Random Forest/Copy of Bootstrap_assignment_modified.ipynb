```python
boston = load_boston()
x=boston.data #independent variables
y=boston.target #target variable
```

```python
def generating_samples(input_data, target_data):
  selecting_rows = list(np.random.choice(len(x), 303,replace= False))
  #print(selecting_rows)
  #print(len(selecting_rows))

  replacing_rows = list(np.random.choice(selecting_rows,size = 203,replace= True))
  #print(replacing_rows)
  #print(len(replacing_rows))

  rand = random.randint(3,13)
  selecting_columns = list(np.random.choice(13,rand,replace= False))
  #print(selecting_columns)

  sample_data = x[selecting_rows][:,selecting_columns]
  
  target_of_sample_data = y[selecting_rows]

  replicated_sample_data = x[replacing_rows][:,selecting_columns]
  target_of_replicated_sample_data = y[replacing_rows]

  final_sample_data = np.vstack((sample_data,replicated_sample_data))

  final_target_data = np.vstack((target_of_sample_data.reshape(-1,1),target_of_replicated_sample_data.reshape(-1,1)))
  

  '''In this function, we will write code for generating 30 samples '''
    # you can use random.choice to generate random indices without replacement
    # Please have a look at this link https://docs.scipy.org/doc/numpy-1.16.0/reference/generated/numpy.random.choice.html for more details
    # Please follow above pseudo code for generating samples 
    

  return final_sample_data , final_target_data,selecting_rows,selecting_columns
    #note please return as lists'''
```

<font color='cyan'> <b> Grader function - 1 </b> </fongt>

```python
def grader_samples(a,b,c,d):
    length = (len(a)==506  and len(b)==506)
    sampled = (len(a)-len(set([str(i) for i in a]))==203)
    rows_length = (len(c)==303)
    column_length= (len(d)>=3)
    assert(length and sampled and rows_length and column_length)
    return True
a,b,c,d = generating_samples(x, y)
grader_samples(a,b,c,d)
```

```python
#Repeating Task for 35


mse_35 = []
oob_35 = []

for task in range(35):
 
  # Creating new datasets
  list_input_data =[]
  list_output_data =[]
  list_selected_row= []
  list_selected_columns=[]

  

  for i in range(0,30):
    a,b,c,d = generating_samples(x,y)
    list_input_data.append(a)
    list_output_data.append(b)
    list_selected_row.append(c)
    list_selected_columns.append(d)

  #print(y.shape)

  list_of_all_models = []

  for i in range(30):
    model = DecisionTreeRegressor(max_depth=None)
    model.fit(list_input_data[i],list_output_data[i])
    filename = "finalized_model{task}{num}.sav".format(task = task,num=i)
    pickle.dump(model, open(filename, 'wb'))
    list_of_all_models.append(filename)
    
  #print(y.shape)

  #Calculating mse
  all_pred = []

  for point in x:
    model_pred = 0
  #print(point,point.shape)
    for i in range(30):
      point_new = point
      point_new = point_new[list_selected_columns[i]]
      filename = "finalized_model{task}{num}.sav".format(task = task,num=i)

      loaded_model = pickle.load(open(filename, 'rb'))
      pred = loaded_model.predict(point_new.reshape(1,-1))
    
      model_pred += pred


    all_pred.append(float(model_pred/30))


  #print(y.shape,'y')
  #print(len(all_pred),'allpred')
  mse_35.append(mean_squared_error(y,all_pred))

  #calculatin oob
  all_pred = []
  for j,point in enumerate(x):
    model_pred = []
    #print(point,point.shape)
    for i in range(30):
      if j  not in list_selected_row[i]:
        point_new = point
        point_new = point_new[list_selected_columns[i]]
        filename = "finalized_model{task}{num}.sav".format(task = task,num=i)

        loaded_model = pickle.load(open(filename, 'rb'))
        pred = loaded_model.predict(point_new.reshape(1,-1))
        #print(pred)
        #print(pred)
      model_pred.append(pred)

    all_pred.append(np.median(model_pred))


  oob_35.append(mean_squared_error(y,all_pred))

  



print(len(oob_35))
print(len(mse_35))


```

```python
#Calculatin CI
#https://stackoverflow.com/questions/15033511/compute-a-confidence-interval-from-sample-data
import numpy as np
import scipy.stats


def mean_confidence_interval(data, confidence=0.95):
    a = 1.0 * np.array(data)
    n = len(a)
    m, se = np.mean(a), scipy.stats.sem(a)
    h = se * scipy.stats.t.ppf((1 + confidence) / 2., n-1)
    return m, m-h, m+h
```

```python
oob_ci = mean_confidence_interval(oob_35)
mse_ci = mean_confidence_interval(mse_35)

print('Mean:','Lower Limit:','Upper_limit:')
print(mse_ci)
print(oob_ci)
```

Observation:

 - In this task we calculated the sample using the 35 random data sample using the function of random sampling then calculted mse and oob scores.

 - Using the score as samples we calculated the 95 % confidence interval for mean of the samples. It means if we repeat the experiment then there is 95% chance that the mean will be observed in the given range.
 

 - For MSE 

  - Mean: 2.3, Upper Range: 2.24, Lowe Range:2.4
